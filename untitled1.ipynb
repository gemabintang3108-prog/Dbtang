```python
from openpyxl import Workbook, load_workbook
import os

data_siswa = []


# --- Fungsi Membaca File Excel Jika Ada ---
def muat_data_dari_excel():
    """Membaca data siswa dari file Excel jika sudah ada"""
    global data_siswa

    if os.path.exists("data_siswa.xlsx"):
        wb = load_workbook("data_siswa.xlsx")
        ws = wb.active

        data_siswa = []
        for row in ws.iter_rows(min_row=2, values_only=True):
            nama, nilai = row
            data_siswa.append([nama, nilai])

        wb.close()
        print("Data lama berhasil dimuat dari 'data_siswa.xlsx'")
    else:
        print("Belum ada file data_siswa.xlsx, mulai dengan data kosong.")


# --- Fungsi Menampilkan Data ---
def tampilkan_data():
    if len(data_siswa) == 0:
        print("\nBelum ada data siswa.")
    else:
        print("\nDaftar Data Siswa:")
        print("=" * 40)
        print(f"{'No':<4} {'Nama':<15} {'Nilai':>6}")
        print("=" * 40)

        for i, s in enumerate(data_siswa, start=1):
            print(f"{i:<4} {s[0]:<15} {s[1]:>6}")

        print("=" * 40)


# --- Fungsi Menambah Data ---
def tambah_data():
    while True:
        nama = input("\nMasukkan nama siswa: ").strip()
        if nama == "":
            print("Nama tidak boleh kosong!")
            continue

        try:
            nilai = int(input("Masukkan nilai (0-100): "))
            if 0 <= nilai <= 100:
                data_siswa.append([nama, nilai])
                print("Data berhasil ditambahkan.")
            else:
                print("Nilai harus antara 0 sampai 100!")
        except ValueError:
            print("Input nilai harus berupa angka!")

        lanjut = input("Tambah data lagi? (y/n): ").lower()
        if lanjut == "n":
            break


# --- Fungsi Mengedit Data ---
def edit_data():
    tampilkan_data()
    if len(data_siswa) == 0:
        return

    try:
        no = int(input("\nMasukkan nomor data yang ingin diedit: "))
        if 1 <= no <= len(data_siswa):

            nama_baru = input(
                f"Masukkan nama baru ({data_siswa[no-1][0]}): "
            ).strip()
            nilai_baru = input(
                f"Masukkan nilai baru ({data_siswa[no-1][1]}): "
            ).strip()

            if nama_baru != "":
                data_siswa[no-1][0] = nama_baru

            if nilai_baru != "":
                try:
                    nilai_baru = int(nilai_baru)
                    if 0 <= nilai_baru <= 100:
                        data_siswa[no-1][1] = nilai_baru
                    else:
                        print("Nilai harus antara 0-100, tidak diubah.")
                except ValueError:
                    print("Input nilai tidak valid, tidak diubah.")

            print("Data berhasil diperbarui.")

        else:
            print("Nomor tidak ditemukan.")
    except ValueError:
        print("Input harus berupa angka.")


# --- Fungsi Menghapus Data ---
def hapus_data():
    tampilkan_data()
    if len(data_siswa) == 0:
        return

    try:
        no = int(input("\nMasukkan nomor data yang ingin dihapus: "))
        if 1 <= no <= len(data_siswa):

            konfirmasi = input(
                f"Yakin ingin menghapus data '{data_siswa[no-1][0]}'? (y/n): "
            ).lower()

            if konfirmasi == "y":
                data_siswa.pop(no - 1)
                print("Data berhasil dihapus.")
            else:
                print("Data batal dihapus.")

        else:
            print("Nomor tidak ditemukan.")
    except ValueError:
        print("Input harus berupa angka.")


# --- Fungsi Mencari Data ---
def cari_data():
    if len(data_siswa) == 0:
        print("\nBelum ada data untuk dicari.")
        return

    keyword = input("\nMasukkan nama siswa yang ingin dicari: ").strip().lower()
    hasil = []

    for i, s in enumerate(data_siswa, start=1):
        if keyword in s[0].lower():
            hasil.append((i, s))

    if len(hasil) == 0:
        print(f"Tidak ditemukan siswa dengan nama '{keyword}'.")
    else:
        print("\nHasil Pencarian:")
        print("=" * 40)
        print(f"{'No':<4} {'Nama':<15} {'Nilai':>6}")
        print("=" * 40)
        for i, s in hasil:
            print(f"{i:<4} {s[0]:<15} {s[1]:>6}")
        print("=" * 40)


# --- Fungsi Menyimpan ke Excel ---
def simpan_ke_excel():
    if len(data_siswa) == 0:
        print("\nBelum ada data untuk disimpan!")
        return

    wb = Workbook()
    ws = wb.active
    ws.title = "Data Siswa"

    ws.append(["Nama", "Nilai"])
    for s in data_siswa:
        ws.append(s)

    wb.save("data_siswa.xlsx")
    print("\nData berhasil disimpan ke file 'data_siswa.xlsx'")


# --- Program Utama ---
muat_data_dari_excel()

while True:
    print("\n=== MENU DATA SISWA ===")
    print("1. Lihat Data Siswa")
    print("2. Tambah Data Siswa")
    print("3. Simpan ke Excel")
    print("4. Keluar")
    print("5. Edit Data Siswa")
    print("6. Hapus Data Siswa")
    print("7. Cari Data Siswa")

    pilih = input("Pilih menu (1-7): ")

    if pilih == "1":
        tampilkan_data()
    elif pilih == "2":
        tambah_data()
    elif pilih == "3":
        simpan_ke_excel()
    elif pilih == "4":
        print("\nTerima kasih! Program selesai.")
        break
    elif pilih == "5":
        edit_data()
    elif pilih == "6":
        hapus_data()
    elif pilih == "7":
        cari_data()
    else:
        print("Pilihan tidak valid! Coba lagi.")
```

```python
import time

print("Masukkan 10 nilai!")

try:
    nilai = []
    for i in range(1, 11):
        masuk = int(input(f"Nilai ke-{i}: "))
        nilai.append(masuk)

    print("Menghitung...")
    time.sleep(1.5)

    rata = sum(nilai) / 10
    print(f"Rata-rata: {rata:.2f}")

    if rata > 100 or rata < 0:
        print("Nilai harus 0–100!")
    elif rata >= 90:
        print("A — Mantap!")
    elif rata >= 80:
        print("B — Good!")
    elif rata >= 70:
        print("C — Lumayan!")
    elif rata >= 60:
        print("D — Tingkatkan!")
    else:
        print("E — Belajar lagi!")

except ValueError:
    print("Input harus angka!")
```
