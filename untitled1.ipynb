```python
nilai = input("Masukkan nilai kamu: ")

try:
    nilai = int(nilai)

    if nilai >= 90:
        print("Grade: A")
    elif nilai >= 80:
        print("Grade: B")
    elif nilai >= 70:
        print("Grade: C")
    elif nilai >= 60:
        print("Grade: D")
    else:
        print("Grade: E")

except ValueError:
    print("Input harus berupa angka!")
```

```python
import time

print("== WARKOP KOKOP ==")
time.sleep(1.5)

print("Masukkan harga barang untuk cek bonus kamu!")
item1 = float(input("Harga barang pertama: "))
item2 = float(input("Harga barang kedua : "))
item3 = float(input("Harga barang ketiga: "))

total = item1 + item2 + item3
print("Total belanja kamu:", total)

if total >= 500000:
    diskon = total * 0.20
    print("Diskon 20% berhasil diterapkan!")
elif total > 300000:
    diskon = total * 0.15
    print("Diskon 15% berhasil diterapkan!")
elif total >= 100000:
    diskon = total * 0.10
    print("Diskon 10% berhasil diterapkan!")
else:
    diskon = 0
    print("Tidak dapat diskon")

print("Bayar setelah diskon:", total - diskon)
```

```python
import time

print("=== WARKOP KOKOP ===")
time.sleep(1)

a = float(input("Harga barang 1: "))
b = float(input("Harga barang 2: "))
c = float(input("Harga barang 3: "))

total = a + b + c

if total >= 500000:
    pot = total * 0.20
    print("Diskon: 20%")
elif total > 300000:
    pot = total * 0.15
    print("Diskon: 15%")
elif total >= 100000:
    pot = total * 0.10
    print("Diskon: 10%")
else:
    pot = 0
    print("Tidak ada diskon")

print("Total setelah diskon:", total - pot)
```

```python
print("Masukkan 3 nilai!")

try:
    a = int(input("Nilai 1: "))
    b = int(input("Nilai 2: "))
    c = int(input("Nilai 3: "))

    rata = (a + b + c) / 3
    print(f"Rata-rata kamu: {rata:.2f}")

    if rata > 100 or rata < 0:
        print("Nilai harus 0–100!")
    elif rata >= 90:
        print("A")
    elif rata >= 80:
        print("B")
    elif rata >= 70:
        print("C")
    elif rata >= 60:
        print("D")
    else:
        print("E")

except ValueError:
    print("Masukkan angka yang valid!")
```

```python
import time

print("Masukkan 10 nilai!")

try:
    nilai = []
    for i in range(1, 11):
        masuk = int(input(f"Nilai ke-{i}: "))
        nilai.append(masuk)

    print("Menghitung...")
    time.sleep(1.5)

    rata = sum(nilai) / 10
    print(f"Rata-rata: {rata:.2f}")

    if rata > 100 or rata < 0:
        print("Nilai harus 0–100!")
    elif rata >= 90:
        print("A — Mantap!")
    elif rata >= 80:
        print("B — Good!")
    elif rata >= 70:
        print("C — Lumayan!")
    elif rata >= 60:
        print("D — Tingkatkan!")
    else:
        print("E — Belajar lagi!")

except ValueError:
    print("Input harus angka!")
```

```python
from openpyxl import load_workbook
import os

data = []

def load_data():
    global data
    if os.path.exists("data_siswa.xlsx"):
        wb = load_workbook("data_siswa.xlsx")
        ws = wb.active

        data = []
        for nama, nilai in ws.iter_rows(min_row=2, values_only=True):
            data.append([nama, nilai])
        wb.close()
        print("Data lama berhasil dimuat!")
    else:
        print("File belum tersedia.")
```
